```python
from PIL import Image
import pytesseract
from googletrans import Translator
from gtts import gTTS
from IPython.display import Audio

# Open an image file (replace with the path to your image)
img = Image.open('/content/drive/MyDrive/OCR/test.png')

# Perform OCR
text = pytesseract.image_to_string(img)

# Print the extracted text
print("Original Text:")
print(text)

# Translate the text into a different language
translator = Translator()
translated_text = translator.translate(text, dest='hi')  # Replace 'fr' with the language code of your choice

# Print the translated text
print("\nTranslated Text:")
print(translated_text.text)

# Convert the translated text to speech
tts = gTTS(translated_text.text, lang='hi')  # Replace 'fr' with the language code of your choice
tts.save('/content/translated_audio.mp3')

# Display the audio
Audio('/content/translated_audio.mp3', autoplay=True)
```
